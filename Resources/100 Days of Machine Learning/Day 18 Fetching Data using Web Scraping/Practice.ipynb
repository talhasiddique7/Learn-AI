# Import Libraries

```python
import numpy as np
import requests
from bs4 import BeautifulSoup
```

# Send Request

```python
webpage=requests.get('https://www.javatpoint.com/cpp-tutorial').text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

# Get div in webpage having class

```python
p=soup.find_all('div',class_='responsivediv')
```

```python
len(p)
```

```python
print(p)
```

# Get p Tags

```python

#Main Work:
tags=[]
for i in p:
    paragraphs = i.find_all('p')
    for paragraph in paragraphs:
        print(paragraph.text)
        
```

# Get 'p' and 'li' tags and store in list

```python
ptag=[]
litag=[]
for i in p:
    a=i.find_all('li')
    z=i.find_all('p')
    for a1 in a:
        litag.append(a1.text)
    for z1 in z:
        ptag.append(z1.text)
    
```

```python
for i in ptag:
    print(i)
```

```python
for i in litag:
    print(i)
```

```python
count=0
for i in p:
    z=i.find_all('p')
    count=count+1
    if count==2:
        break
    for a in z:
        c=i.find_all('li')
        print(a.text)
        print('\n')
        for w in c:
            print(w.text)
        print('\n')
```
